```
#Bibliotecas
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score
```

```
'''
Entrada de dados
features (1 sim, 0 não)
pelo longo? 
perna curta?
late?

'''

porco1 = [0, 1, 0]
porco2 = [0, 1, 1]
porco3 = [1, 1, 0]

cachorro1 = [0, 1, 1]
cachorro2 = [1, 0, 1]
cachorro3 = [1, 1, 1]
```

```
# 1 => porco, 0 => cachorro
teste_x = [porco1, porco2, porco3, cachorro1, cachorro2, cachorro3]
teste_y = [1,1,1,0,0,0]
```

```
#A variável 'model' será o estimador para análise dos dados/ o método fit() receberá o array 'dados' e 'classes' como parâmetros para realizar o aprendizado supervisionado 
model = LinearSVC()
model.fit(teste_x, teste_y)
```

```
#Testando o modelo
animal_desconhecido = [1,1,1]
model.predict([animal_desconhecido])
```

```
#Gerando dados para predição
desconhecido1 = [1,1,1]
desconhecido2 = [1,1,0]
desconhecido3 = [0,1,1]
#A variável 'testes' receberá os dados/ 'testes_classes' receberá o valor real previamente conhecido
teste_x = [desconhecido1, desconhecido2, desconhecido3]
teste_y = [0, 1, 1]
previsoes = model.predict(teste_x)
```

```
#Utilizando o método accuracy_score do scikit learn para cacular a taxa de acerto do modelo
taxa_de_acerto = accuracy_score(teste_y, previsoes)
print("Taxa de acerto %.2f " % (taxa_de_acerto * 100)+"%")
```
