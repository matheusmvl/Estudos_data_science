```
#bibliotecas
import pandas as pd
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```
#utilizando a biblioteca Pandas para visualização dos arquivos csv através da variável dados
uri='https://gist.githubusercontent.com/guilhermesilveira/2d2efa37d66b6c84a722ea627a897ced/raw/10968b997d885cbded1c92938c7a9912ba41c615/tracking.csv'
dados = pd.read_csv(uri)
```

```
#renomeando as colunas para adequação do projeto à lingua portuguesa atravpes do dicionário 'mapa'
mapa = { 
    "home" : "principal",
    "how_it_works" : "como_funciona",
    "contact" : "contato",
    "bought" : "comprou"
}
dados = dados.rename(columns = mapa)
#com dados.shape verificamos quantos elementos temos nos dados e o formato deles
dados.shape
```

```
#adequando as variáveis para criação do modelo
x = dados[["principal","como_funciona","contato"]]
y = dados[["comprou"]]
```

```
#separando dados para treino e teste utilizando o sklearn
SEED = 20
treino_x, teste_x, treino_y, teste_y = train_test_split(x, y,
                                                        random_state = SEED,
                                                        test_size = 0.25,
                                                        stratify = y)
```

```
#criação do modelo
modelo = LinearSVC()
modelo.fit(treino_x, treino_y)
previsoes = modelo.predict(teste_x)
#calculando a acurácia do modelo
taxa_de_acerto = accuracy_score(teste_y, previsoes) * 100
print("A taxa de acerto foi %.2f%%" % taxa_de_acerto)
```
